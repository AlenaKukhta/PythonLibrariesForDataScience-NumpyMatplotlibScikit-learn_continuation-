# Урок 2. Анализ данных и проверка статистических гипотез.

## 1. В чём различие между зависимыми и независимыми выборками?

Зависимыми являются выборки, если каждому случаю из одной выборки соответствует один и только один случай из другой выборки и наоборот. 

Иначе выборки являются независимыми.

Таким образом, у зависимых выборок всегда одинаковый объем, объем независимых может отличаться.

## 2. Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

Непараметрические методы наиболее приемлемы, когда объем выборок мал или информация о выборке отсутствует, а параметрические - если данных много (например, n > 100), поскольку:

1) у очень больших выборок выборочные средние подчиняются нормальному закону, даже если исходная переменная не является нормальной или измерена с погрешностью;

2) непараметрические тесты имеют меньшую статистическую мощность (менее чувствительны), чем параметрические (могут обнаружить даже слабые отклонения).
