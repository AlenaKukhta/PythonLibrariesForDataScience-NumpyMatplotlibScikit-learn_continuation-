# Урок 1. Введение в задачу классификации. Постановка задачи и подготовка данных.

## Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

**Precision** - точность - отражает какой процент объектов, отнесённых к "основному" классу, действительно являются таковыми. Преобладающее использование метода позволяет получить бОльшую уверенность в прогнозе. Следовательно экономию от реализации проекта. Например:

1) Моделирование в одежде: по совокупности измерений тела определяется тип фигуры и подходящие ему модели одежды;

2) Предсказания катастроф (например, земной коре (извержение вулкана, землетрясение), гидросфере (цунами, наводнение, лимнологическая катастрофа), атмосфере (ураган, тайфун, смерч, буран, озоновая дыра), магнитосфере и т.д.);

3) Диагностика в медицине: по совокупности симптомов, исследований и результатов лабораторных анализов (во избежении передиагностики и исключения залечивания человека при отсутствии заболевания);

4) Разведка месторождений полезных ископаемых: по данным геологического строения месторождения; по данным пространственного расположения, условий залегания, форм, размеров и строений залежей; по данным количества и качества полезных ископаемых; по данным технологических свойств залежей и факторов, определяющих условия эксплуатации месторождения (точность позволит минимизировать затраты по добыче полезных ископаемых в местах их отсутствия);

5) Оценка покупательской способности, например, при определении места расположения торговой точки.

**Recall** - полнота - отражает какой процент объектов "основного" класса было обнаружено алгоритмом. Преобладающее использование метода позволяет определить наибольшее количество положительных результатов. Это важно при принятии стратегических решений (готовься к худшему - надейся на лучшее). Например:

1) Выявление мошеннических действий;

2) Оценке рисков (например, невыплат при кредитовании, налоговых, судебных и т.д.);

3) Маркетинговое исследование, например, при определении товарного ассортимента.

## Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

*F-score* - метрика объединяющая в себе точность и полноту, выдавая как результат их гармоническое среднее, а коэффициент $\beta$, позволяет балансирвоать вклад, вносимый точностью, тем самым акцентируя внимание на той метрике, максимизация которой для нас наиболее приоритетна.

Среднее же от Precision и Recall всегда статично, нивилирует их вес.
